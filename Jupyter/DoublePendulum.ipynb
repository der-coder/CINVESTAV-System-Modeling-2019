# Double Pendulum

> In physics and mathematics, in the area of dynamical systems, a double pendulum is a pendulum with another pendulum attached to its end, and is a simple physical system that exhibits rich dynamic behavior with a strong sensitivity to initial conditions.

[From Wikipedia](https://en.wikipedia.org/wiki/Double_pendulum)

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from math import sin, cos
```

```python
def feval(funcName, *args):
    return eval(funcName)(*args)
```

## Functions

```python
def doublePendulum(t, x):
    dx = np.zeros((len(x)))
        
    theta = x[0]
    d_theta = x[1]
    alpha = x[2]
    d_alpha = x[3]
    
    A = np.array([[3+2*cos(alpha),1+cos(alpha)],[1+cos(alpha),1]])
    b = np.array([[2*d_theta*d_alpha*sin(alpha)-(d_alpha^2)*sin(alpha)],[(d_theta^2)*sin(alpha)]])
    
    x_sol = np.matmul(np.linalg.inv(A),-b)
    
    dx[0]= d_theta
    dx[1] = x_sol[0]
    dx[2]= d_alpha
    dx[3] = x_sol[1]
    
    return dx
```

```python
def doublePendulumGravity(t,x):
    dx = np.zeros((len(x)))
    
    g=9.81
        
    theta = x[0]
    d_theta = x[1]
    alpha = x[2]
    d_alpha = x[3]
    
    A = np.array([[3+2*cos(alpha),1+cos(alpha)],[1+cos(alpha),1]])
    b = np.array([[sin(alpha)*(-2*d_theta*d_alpha-g*cos(theta)-(d_alpha^2))+sin(theta)*(2*g+g*cos(alpha))],[(d_theta^2)*sin(alpha)-g*sin(theta-alpha)]])
    
    x_sol = np.matmul(np.linalg.inv(A),-b)
    
    dx[0]= d_theta
    dx[1] = x_sol[0]
    dx[2]= d_alpha
    dx[3] = x_sol[1]
    
    return dx
```

## Initial Conditions

```python
x0 = np.array([1.0,1.0,1.0,1.0])
```

## Solve the system numerically

```python
tempSol = integrate.solve_ivp(doublePendulum,[0,10],x0, max_step=0.05)
tempSol2 = integrate.solve_ivp(doublePendulumGravity,[0,10],x0, max_step=0.05)
```

## Plot your results

### Double Pendulum without gravity

```python
xs = np.transpose(tempSol.y)
ts = np.transpose(tempSol.t)

plt.figure(num=1,figsize=(20,10))
plt.plot(ts, xs[:,0],ts,xs[:,1])
plt.xlim(0, 12)
plt.legend(['theta','d_theta'], loc=5,fontsize=20,frameon=False)
plt.xlabel('Time', fontsize=17)
plt.tick_params(labelsize='large')
plt.ylabel('Solution', fontsize=17)
plt.title("Plot against time for theta and d_theta",fontsize=24)
plt.tight_layout()
    
plt.figure(num=2,figsize=(20,10))
plt.plot(ts, xs[:,2], ts,xs[:,3])
plt.xlim(0, 12)
plt.legend(["alpha","d_alpha"], loc=5,fontsize=20,frameon=False)
plt.xlabel('Time', fontsize=17)
plt.ylabel('Solution', fontsize=17)
plt.title("Plot against time for alpha and d_alpha",fontsize=24)
plt.tight_layout()

plt.figure(num=3,figsize=(20,10))
plt.plot(xs[:,0], xs[:,1])
#plt.legend(["Posicion angular alpha"], loc=5,fontsize=20,frameon=False)
plt.xlabel('theta', fontsize=17)
plt.ylabel('d_theta', fontsize=17)
plt.title("Phase plot for theta and d_theta",fontsize=24)
plt.tight_layout()
    
plt.figure(num=4,figsize=(20,10))
plt.plot(xs[:,2], xs[:,3])
#plt.xlim(0, 12)
#plt.legend(["Velocidad Angular alpha"], loc=5,fontsize=20,frameon=False)
plt.xlabel('alpha', fontsize=17)
plt.ylabel('d_alpha', fontsize=17)
plt.title("Phase plot for alpha and d_alpha",fontsize=24)
plt.tight_layout()
```
