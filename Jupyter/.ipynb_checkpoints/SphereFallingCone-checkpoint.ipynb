# Import libraries
Code made in Jupyter using the SageMath 8.9 kernel

```python
import numpy as np # Linar Algrebra Operations
from scipy import integrate # ODE solvers
import math as mt # Math functions
import matplotlib.pyplot as plt # Graph functions
```

## Problem \# 01

### Initial Conditions

```python
m= 1; # kilogram
g = 9.81 # meter/second^2
t = [0,10] # Evaluation time
x0_ip = np.array([3.0,0.0,1.0,1.0]) # Initial conditions
f_pos = 0.0;
f_theta = 0.0;
alpha = 1.0; #radians
```

### Equations of the system

```python
def sphereFall(t, x):
    dx = np.zeros((len(x))) # Create the matrix to store the values
        
    pos = x[0]  # Define state variables
    d_pos = x[1]
    theta = x[2]
    d_theta = x[3]

    # Define matrix of the syste
    # Such that A d2_q + b = forces
    
    A = np.array([[m*((1/mt.sin(alpha))^2), 0], # Python math library does not have a csc function
                  [0, m*(pos^2)]])

    b = np.array([[-m*pos*(d_theta)^2 + m*g/mt.tan(alpha)], # Python math library does not have a cot function
                  [2*m*pos*d_theta*d_pos]])
        
    forces = np.array([[f_pos],[f_theta]])
    
    x_sol = np.matmul(np.linalg.inv(A),forces-b)
    
    dx[0]= d_pos
    dx[1] = x_sol[0]
    dx[2]= d_theta
    dx[3] = x_sol[1]
    
    return dx
```

```python
def sphereHamilton(t, x):
    dx = np.zeros((len(x))) # Create the matrix to store the values
        
    r = x[0]  # Define state variables
    theta = x[1]
    p_r = x[2]
    p_theta = x[3]
 
    dx[0]= p_r/(m*(1/mt.sin(alpha))^2)
    dx[1] = p_theta/(m*(r)^2)
    dx[2]= (((p_theta)^2)/(m*(r)^3))-(m*g/mt.tan(alpha))
    dx[3] = 0
    
    return dx
```

Evaluate the function and store the values. Scipy's solve_ivp function will use RK45 by default.

```python
results = integrate.solve_ivp(sphereFall,[0,10],x0_ip,  max_step=0.05)
```

Plot the results.

## Euler-Lagrange

```python
xs = np.transpose(results.y)
ts = np.transpose(results.t)


plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.figure(num=1,figsize=(12,8))
plt.plot(ts, xs[:,0],"--k",ts,xs[:,2],"k", linewidth=4)
plt.xlim(0, 10)
plt.legend([u'$r$',u'$\\theta$'], fontsize=24,frameon=False, bbox_to_anchor=(0.5,-0.2), loc="center", mode="expand", ncol=2)
#plt.legend(bbox_to_anchor=(0.2,-0.2), loc="lower left")
plt.xlabel(u'Tiempo', fontsize=24)
plt.tick_params(labelsize='20')
plt.ylabel(u"Soluci\\'on", fontsize=24)
plt.title(u"Comparaci\\'on de $r$ y $\\theta$",fontsize=40)
plt.tight_layout()

plt.figure(num=2,figsize=(12,8))
plt.plot(ts, xs[:,1],"--k",ts,xs[:,3],"k", linewidth=4)
plt.xlim(0, 10)
plt.legend([u'$\dot{r}$',u'$\dot{\\theta}$'], fontsize=24,frameon=False, bbox_to_anchor=(0.5,-0.2), loc="center", mode="expand", ncol=2)
plt.xlabel(u'Tiempo', fontsize=24)
plt.tick_params(labelsize='18')
plt.ylabel(u"Soluci\\'on", fontsize=24)
plt.title(u"Comparaci\\'on de $\dot{r}$ y $\dot{\\theta}$",fontsize=40)
plt.tight_layout()
    
    
plt.figure(num=3,figsize=(12,8))
plt.plot(xs[:,0],xs[:,1],"k", linewidth=4)
plt.xlabel(u'$r$', fontsize=24)
plt.tick_params(labelsize='18')
plt.ylabel(u"$\dot{r}$", fontsize=24)
plt.title(u"Diagrama fase de $r$ y $\dot{r}$",fontsize=40)
plt.tight_layout()

plt.figure(num=4,figsize=(12,8))
plt.plot(xs[:,2],xs[:,3],"k", linewidth=4)
plt.xlabel(u'$\\theta$', fontsize=24)
plt.tick_params(labelsize='18')
plt.ylabel(u"$\dot{\\theta}$", fontsize=24)
plt.title(u"Diagrama fase de $\\theta$ y $\dot{\\theta}$",fontsize=40)
plt.tight_layout()  
```
